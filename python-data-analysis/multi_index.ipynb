## The .pivot() Method

```python
df = pd.read_csv('data/pandas/salesmen.csv',parse_dates=['Date'])
df['Salesman'] = df['Salesman'].astype('category')
df.head()
```

```python
df.pivot(index='Date',columns='Salesman',values='Revenue').head()
```

## The .pivot_table() Method

```python
df = pd.read_csv('data/pandas/foods.csv')
df.head()
```

```python
df.pivot_table(values='Spend',index='Gender',aggfunc='mean')
```

```python
df.pivot_table(values='Spend',index='Gender',aggfunc='sum')
```

```python
df.pivot_table(values='Spend',index='Gender',aggfunc='count')
```

```python
df.pivot_table(values='Spend',index='City',aggfunc='mean')
```

```python
df.pivot_table(values='Spend',index='Item',aggfunc='sum')
```

```python
df.pivot_table(values='Spend',index=['Gender','Item'],aggfunc='sum')
```

```python
df.pivot_table(values='Spend',index=['Gender','Item'],columns='City',aggfunc='mean')
```

```python
df.pivot_table(values='Spend',index=['Gender','Item'],columns=['Frequency','City'],aggfunc='mean')
```

```python
df.pivot_table(values='Spend',index=['Gender','Item'],columns='City',aggfunc='max')
```

```python
pd.pivot_table(data=df,values='Spend',index=['Gender','Item'],columns='City',aggfunc='max')
```
