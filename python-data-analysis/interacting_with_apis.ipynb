# Interacting with Web APIs

```python
import numpy as np
import pandas as pd
import requests
```

```python
# github url for repository issues
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
# get response
resp = requests.get(url)
resp
```

```python
# return json data from the first issue
data = resp.json()
data[0]['title']
```

```python
# create a dataframe of issues
issues = pd.DataFrame(data, columns=['number','title','labels','state'])
issues
```
