# Tree-Based Methods

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/kyphosis.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df, hue='Kyphosis')
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

## Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/loan_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(df[df['credit.policy']==1]['fico'],
             bins=35,
             kde=False,
             label='Credit Policy = 1')
sns.distplot(df[df['credit.policy']==0]['fico'],
             bins=35,
             kde=False,
             label='Credit Policy = 0')
plt.legend()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(df[df['not.fully.paid']==1]['fico'],
             bins=35,
             kde=False,
             label='Not Fully Paid= 1')
sns.distplot(df[df['not.fully.paid']==0]['fico'],
             bins=35,
             kde=False,
             label='Not Fully Paid = 0')
plt.legend()
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x='purpose', data=df, hue='not.fully.paid', palette='Set1')
```

```python
sns.jointplot(x='fico', y='int.rate', data=df, color='purple')
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(x='int.rate', y='fico', data=df, hue='credit.policy', col='not.fully.paid', palette='Set1')
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(df, columns=cat_feats, drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train, y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```
