```python
def indeed_query(q,n,l):
    '''
    workaround function to override the 10 request limit. 
    Takes in a query, number of results, and a location.
    '''

    appended_data = []
    
    for start in range(0,n,10):
        # query parameters
        params = {
        'q' : q,
        'l' : l,
        'format': 'json',
        'radius' : 5,
        'latlong': 1,
        'userip' : "1.2.3.4",
        'useragent' : "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_2)",
        'start': start
        }

        # preform search
        search_response = client.search(**params)

        # normalize search results
        df = json_normalize(search_response['results'])

        appended_data.append(df)
    
    # concatenate dataframes
    appended_data = pd.concat(appended_data, axis=0)
    
    return appended_data
    
    
```

```python
austin = indeed_query('data analyst',100,'Austin, TX')
boston = indeed_query('data analyst',100,'Boston, MA')
ny = indeed_query('data analyst',100,'New York, NY')
phoenix = indeed_query('data analyst',100,'Phoenix, AZ')
```

```python
cities = [austin,boston,ny,phoenix]
jobs = pd.concat(cities, axis=0)
```
